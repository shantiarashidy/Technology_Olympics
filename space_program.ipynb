```python
df['Proficiency_Rate'] = df['Proficient'] / df['Total']
df.head()
```

```python
math_df = df[df['Topic'] == 'Math'].copy()
reading_df = df[df['Topic'] == 'Reading'].copy()

math_df.head()
```

```python
# For Math: Count unique school years per grade
math_grade_years = math_df.groupby('Grade')['School Year'].nunique()

# Show which grades have how many years of data
print("Math - Years of data per grade:")
print(math_grade_years)

# Filter to keep only grades with 2+ years
math_allowed_grades = math_grade_years[math_grade_years >= 2].index.tolist()
print(f"\nAllowed Math grades: {math_allowed_grades}")
```

```python
# Filter math_df to only include allowed grades
math_allowed = math_df[math_df['Grade'].isin(math_allowed_grades)]

# Group by Grade AND District, then calculate std of proficiency rates
math_stability = math_allowed.groupby(['Grade', 'District'])['Proficiency_Rate'].std()

# Show first few rows
print("Stability scores (standard deviation) for Math:")
print(math_stability.head(10))
```

```python
# For each grade, find the district with minimum std
# We use idxmin() to get the index of minimum value

# First, reset index to make manipulation easier
math_stability_df = math_stability.reset_index()

# For each grade, get the row with minimum std
math_most_stable = math_stability_df.loc[
    math_stability_df.groupby('Grade')['Proficiency_Rate'].idxmin()
]

# Keep only columns we need and sort by grade
stable_math_districts = math_most_stable[['Grade', 'District']].sort_values('Grade')

# Reset index for clean output
stable_math_districts = stable_math_districts.reset_index(drop=True)

print("Most stable Math districts per grade:")
print(stable_math_districts)
```

```python
# Repeat the same process for Reading
reading_grade_years = reading_df.groupby('Grade')['School Year'].nunique()
reading_allowed_grades = reading_grade_years[reading_grade_years >= 2].index.tolist()

# Filter and calculate stability
reading_allowed = reading_df[reading_df['Grade'].isin(reading_allowed_grades)]
reading_stability = reading_allowed.groupby(['Grade', 'District'])['Proficiency_Rate'].std()

# Find most stable per grade
reading_stability_df = reading_stability.reset_index()
reading_most_stable = reading_stability_df.loc[
    reading_stability_df.groupby('Grade')['Proficiency_Rate'].idxmin()
]

# Format output
stable_reading_districts = reading_most_stable[['Grade', 'District']].sort_values('Grade')
stable_reading_districts = stable_reading_districts.reset_index(drop=True)

print("Most stable Reading districts per grade:")
print(stable_reading_districts)
```

```python
# What is the latest school year in our data?
latest_year = df['School Year'].max()
print(f"The most recent school year is: {latest_year}")

# Let's see what other years exist to confirm
all_years = sorted(df['School Year'].unique())
print(f"All school years: {all_years}")
```

```python
# Keep only data from the latest school year
latest_data = df[df['School Year'] == latest_year].copy()

print(f"Total rows in latest year: {len(latest_data)}")
print(f"Unique districts in latest year: {latest_data['District'].nunique()}")
print(f"Topics included: {latest_data['Topic'].unique()}")
```

```python
# Look at a sample from the latest year
print("Sample from latest year:")
print(latest_data[['Topic', 'Grade', 'District', 'Proficient', 'Total']].head(10))

# Quick check: Are Proficient and Total numbers or already aggregated?
print(f"\nProficient column type: {latest_data['Proficient'].dtype}")
print(f"\nExample: A district with 37.0 proficient out of 61.0 total")
print(f"This means 37 students passed out of 61 tested")
```

```python
# Group by district and sum the proficient column
district_proficient = latest_data.groupby('District')['Proficient'].sum()

# Let's see what we got
print("Proficient students per district (first 10):")
print(district_proficient.head(10))

# How many districts do we have?
print(f"\nTotal districts with data in {latest_year}: {len(district_proficient)}")
```

```python
# Calculate grand total of all proficient students
total_proficient_nationwide = district_proficient.sum()

print(f"Total proficient students in {latest_year}: {total_proficient_nationwide:.0f}")
```

```python
# Calculate probability for each district
district_probability = district_proficient / total_proficient_nationwide

print("Probability for each district (first 10):")
print(district_probability.head(10))

# Let's verify: probabilities should sum to 1
print(f"\nSum of all probabilities: {district_probability.sum():.4f}")
```

```python
# Sort probabilities in descending order and take top 10
top_10_probabilities = district_probability.sort_values(ascending=False).head(10)

print("Top 10 districts by probability:")
print(top_10_probabilities)
```

```python
# Create the final DataFrame
most_probable = pd.DataFrame({
    'District': top_10_probabilities.index,
    'Probability': top_10_probabilities.values
})

# Round to 3 decimal places as required
most_probable['Probability'] = most_probable['Probability'].round(3)

# It's already sorted from step 7, but let's be explicit
most_probable = most_probable.sort_values('Probability', ascending=False)

# Reset index for clean output
most_probable = most_probable.reset_index(drop=True)

print("Final output for Part 2:")
print(most_probable)
print(f"\nNumber of rows: {len(most_probable)} (should be 10)")
```

با استفاده از یک آزمون آماری مناسب، ۱۵ منطقه برتر آموزشی در دو سال اخیر تحصیلی را که بیشترین پیشرفت معنادار را داشته‌اند و در متغیر `significant` ذخیره کنید.

```python
# Find the two most recent school years
all_years = sorted(df['School Year'].unique(), reverse=True)  # descending order
print(f"All years (descending): {all_years}")

year2 = all_years[0]  # Most recent
year1 = all_years[1]  # Second most recent

print(f"\nComparing: {year1} vs {year2}")
print(f"Year 1 (older): {year1}")
print(f"Year 2 (newer): {year2}")

# Find the highest grade
highest_grade = df['Grade'].max()
print(f"\nHighest grade in dataset: {highest_grade}")
```

```python
# Filter for Math, highest grade, and our two years
part3_data = df[
    (df['Topic'] == 'Math') & 
    (df['Grade'] == highest_grade) & 
    (df['School Year'].isin([year1, year2]))
].copy()

print(f"Filtered data shape: {part3_data.shape}")
print(f"\nSample of filtered data:")
print(part3_data.head(10))

# Let's see what districts we have
print(f"\nUnique districts: {part3_data['District'].nunique()}")
```

```python
# For each district, check if they have data in both years
district_year_counts = part3_data.groupby('District')['School Year'].nunique()

# Districts with data in both years
districts_with_both = district_year_counts[district_year_counts == 2].index.tolist()

print(f"Districts with data in both {year1} and {year2}: {len(districts_with_both)}")
print(f"Districts with only one year: {district_year_counts[district_year_counts == 1].count()}")
```

```python
# Filter to only districts with both years
part3_complete = part3_data[part3_data['District'].isin(districts_with_both)]

# Separate into two years
data_year1 = part3_complete[part3_complete['School Year'] == year1]
data_year2 = part3_complete[part3_complete['School Year'] == year2]

print(f"Year {year1}: {len(data_year1)} rows")
print(f"Year {year2}: {len(data_year2)} rows")

# Merge them so we have one row per district with both years
comparison_df = data_year1[['District', 'Proficient', 'Total']].merge(
    data_year2[['District', 'Proficient', 'Total']], 
    on='District', 
    suffixes=('_year1', '_year2')
)

print(f"\nComparison dataframe shape: {comparison_df.shape}")
print("\nFirst few rows:")
print(comparison_df.head())
```

```python
# Calculate rates for each year
comparison_df['rate_year1'] = comparison_df['Proficient_year1'] / comparison_df['Total_year1']
comparison_df['rate_year2'] = comparison_df['Proficient_year2'] / comparison_df['Total_year2']

# Calculate the improvement (difference in rates)
comparison_df['improvement'] = comparison_df['rate_year2'] - comparison_df['rate_year1']

print("Districts with their improvement:")
print(comparison_df[['District', 'rate_year1', 'rate_year2', 'improvement']].head(10))

# Calculate pooled proportion
comparison_df['p_pooled'] = (
    (comparison_df['Proficient_year1'] + comparison_df['Proficient_year2']) / 
    (comparison_df['Total_year1'] + comparison_df['Total_year2'])
)

# Calculate standard error
comparison_df['std_error'] = np.sqrt(
    comparison_df['p_pooled'] * 
    (1 - comparison_df['p_pooled']) * 
    (1/comparison_df['Total_year1'] + 1/comparison_df['Total_year2'])
)

# Calculate z-score
comparison_df['z_score'] = comparison_df['improvement'] / comparison_df['std_error']

print("\nDistricts with z-scores:")
print(comparison_df[['District', 'improvement', 'z_score']].head(10))
```

```python
# Filter for districts with positive improvement (they actually got better)
improved_districts = comparison_df[comparison_df['improvement'] > 0].copy()

print(f"Districts with positive improvement: {len(improved_districts)}")
print(f"Districts with negative or zero improvement: {len(comparison_df) - len(improved_districts)}")

# Sort by z-score (highest = most significant improvement)
improved_districts = improved_districts.sort_values('z_score', ascending=False)

print("\nTop improvements by z-score:")
print(improved_districts[['District', 'rate_year1', 'rate_year2', 'improvement', 'z_score']].head(15))
```

```python
# Take top 15 districts by z-score
top_15_significant = improved_districts.head(15)

# Create the output list as required
significants = top_15_significant['District'].tolist()

print("Top 15 districts with most significant improvement:")
print(significants)
print(f"\nLength: {len(significants)} (should be 15)")
```
