```python
df_train = df_train.drop(columns=["Cement Moisture Factor"])
df_test = df_test.drop(columns=["Cement Moisture Factor"])
```

```python
def remove_outliers(df, col):
    q1, q3 = np.percentile(df[col], [25, 75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return df[(df[col] >= lower) & (df[col] <= upper)]

clean_df = df_train.copy()

for col in ["Water", "Cement Impurity Factor", "cement per water"]:
    clean_df = remove_outliers(clean_df, col)
```

```python
X_train = clean_df.drop(columns=["Strength"])
y_train = clean_df["Strength"]

X_test = df_test.copy()
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazirmatn;font-size:medium">
<font face="vazirmatn" size=3>
در این بخش، وظیفه شما توسعه مدلی است که با استفاده از داده‌های آموزش و بدون دسترسی به برچسب‌ها (labels) بتواند تشخیص دهد که هر نمونه از داده‌های آزمون، یک نمونه عادی است یا یک نمونه ناهنجار.
این مدل باید تنها با تکیه بر الگوها و توزیع ویژگی‌ها در داده‌های آموزش، رفتار نرمال سیستم را یاد بگیرد و بر اساس آن، میزان ناهنجاری هر نمونه از داده‌های آزمون را تخمین بزند.
</font>

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=500,
    max_depth=20,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
df_test.to_csv("submission.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split

X = clean_df.drop(columns=["Strength"])
y = clean_df["Strength"]

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.15, random_state=42
)
```

```python
import matplotlib.pyplot as plt
import numpy as np

importances = model.feature_importances_
indices = np.argsort(importances)[::-1]
features = X.columns

plt.figure(figsize=(10,5))
plt.bar(range(len(features)), importances[indices])
plt.xticks(range(len(features)), features[indices], rotation=45)
plt.title("Feature Importances")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import r2_score

valid_pred = model.predict(X_valid)
r2 = r2_score(y_valid, valid_pred)
print("Validation R2:", r2)
```

```python
final_X = clean_df.drop(columns=["Strength"])
final_y = clean_df["Strength"]

final_model = RandomForestRegressor(
    n_estimators=600,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features="sqrt",
    bootstrap=True,
    random_state=42,
    n_jobs=-1
)

final_model.fit(final_X, final_y)
```

```python
X_test = df_test[final_X.columns]
```

```python
df_test["Strength"] = final_model.predict(X_test)
```

```python
import zipfile
import joblib
import os

if not os.path.exists(os.path.join(os.getcwd(), 'notebook.ipynb')):
    %notebook -e notebook.ipynb

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

submission.to_csv('submission.csv', index=False)
file_names = ['notebook.ipynb', 'submission.csv']
compress(file_names)
```
