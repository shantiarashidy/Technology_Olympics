```python
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import IsolationForest
from sklearn.covariance import EllipticEnvelope
from sklearn.neighbors import LocalOutlierFactor
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
# Separate features from your engineered DataFrames
# We have 15 features total: 6 original + 9 engineered

# List all feature columns we'll use
feature_columns = [
    'Temperature', 'Pressure', 'Vibration', 'Flow Rate', 
    'Humidity', 'Rotation Speed',
    'Temperature/Pressure', 'Vibration/Rotation Speed',
    'Temperature. Humidity', 'Pressure. Flow Rate',
    'Vibration * Rotation Speed', 'Flow rate / pressure',
    'Humidity / Temprature ', 'Total system energy throughput',
    'Centrifugal force proportional to ω²'
]

# Create feature matrices
X_train = df_train[feature_columns].copy()
X_test = df_test[feature_columns].copy()

print(f"Training features shape: {X_train.shape}")
print(f"Test features shape: {X_test.shape}")
print(f"\nFeature names:\n{feature_columns}")
```

```python
# Check for any issues in our features before modeling
print("=== Data Quality Check ===")

# Check for missing values
print(f"Missing values in training: {X_train.isnull().sum().sum()}")
print(f"Missing values in test: {X_test.isnull().sum().sum()}")

# Check for infinite values
print(f"Infinite values in training: {np.isinf(X_train).sum().sum()}")
print(f"Infinite values in test: {np.isinf(X_test).sum().sum()}")

# Check feature scales (to confirm we need scaling)
print("\n=== Feature Scale Ranges (min - max) ===")
for col in feature_columns:
    train_range = X_train[col].max() - X_train[col].min()
    test_range = X_test[col].max() - X_test[col].min()
    print(f"{col[:30]:<30} Train: {X_train[col].min():.2f} - {X_train[col].max():.2f} | Test: {X_test[col].min():.2f} - {X_test[col].max():.2f}")
```

```python
# Split training data: 90% for actual training, 10% for validation
# We assume all training data is "normal" (label = 0)
# This helps us verify our model doesn't flag too many false positives

X_train_model, X_val = train_test_split(
    X_train, 
    test_size=0.10,  # 10% for validation
    random_state=42
)

print(f"Model training set: {X_train_model.shape[0]} samples")
print(f"Validation set: {X_val.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")
print("\nNote: Validation set is assumed to be all normal (label=0)")
print("We expect our model to flag very few anomalies here (<1%)")
```

```python
# Apply RobustScaler to handle different feature scales
# Why RobustScaler over StandardScaler?
# - Uses median and IQR instead of mean and std
# - Less influenced by outliers in training data
# - Better for industrial sensor data with potential anomalies

scaler = RobustScaler()

# Fit scaler ONLY on training data, then transform everything
X_train_scaled = scaler.fit_transform(X_train_model)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# Convert back to DataFrame for easier inspection
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=feature_columns)
X_val_scaled_df = pd.DataFrame(X_val_scaled, columns=feature_columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=feature_columns)

print("=== After RobustScaler ===")
print("Training data statistics (should have median~0, IQR~1):")
print(X_train_scaled_df.describe().loc[['mean', 'std', 'min', '25%', '50%', '75%', 'max']])
```

```python
# Train Isolation Forest - our primary anomaly detector
# This algorithm works by randomly partitioning the data
# Anomalies are isolated with fewer partitions (shorter path length)

# Start with conservative contamination estimate
# We'll validate this against our validation set
iso_forest = IsolationForest(
    n_estimators=150,        # Number of trees in the forest
    contamination=0.05,      # Expected proportion of anomalies (start with 5%)
    random_state=42,         # For reproducibility
    max_samples=256,         # Subset size for each tree
    bootstrap=False,         # Use all samples for each tree
    n_jobs=-1               # Use all CPU cores
)

# Train the model
iso_forest.fit(X_train_scaled)

print("Isolation Forest trained successfully!")
print(f"Model parameters:")
print(f"- Trees: {iso_forest.n_estimators}")
print(f"- Contamination: {iso_forest.contamination}")
print(f"- Features used: {X_train_scaled.shape[1]}")
```

```python
# Get anomaly scores for all datasets
# In Isolation Forest:
# - decision_function returns anomaly scores
# - Negative scores = anomalies (lower = more anomalous)
# - We convert to positive scores where HIGHER = more anomalous

# Get raw scores
train_scores = -iso_forest.decision_function(X_train_scaled)
val_scores = -iso_forest.decision_function(X_val_scaled)
test_scores = -iso_forest.decision_function(X_test_scaled)

# Normalize scores to 0-1 range for easier interpretation
from sklearn.preprocessing import MinMaxScaler
score_scaler = MinMaxScaler()

# Fit on training scores, transform all
train_scores_norm = score_scaler.fit_transform(train_scores.reshape(-1, 1)).ravel()
val_scores_norm = score_scaler.transform(val_scores.reshape(-1, 1)).ravel()
test_scores_norm = score_scaler.transform(test_scores.reshape(-1, 1)).ravel()

print("=== Score Statistics ===")
print(f"Training scores - Mean: {train_scores_norm.mean():.3f}, Max: {train_scores_norm.max():.3f}")
print(f"Validation scores - Mean: {val_scores_norm.mean():.3f}, Max: {val_scores_norm.max():.3f}")
print(f"Test scores - Mean: {test_scores_norm.mean():.3f}, Max: {test_scores_norm.max():.3f}")

# Check validation set - what % would be flagged as anomalous at different thresholds?
print("\n=== Validation Set Analysis (Expected: Near 0% anomalies) ===")
for threshold in [0.7, 0.8, 0.9, 0.95]:
    anomalies = (val_scores_norm > threshold).sum()
    print(f"Threshold {threshold}: {anomalies} anomalies ({anomalies/len(val_scores_norm)*100:.2f}%)")
```

```python
# Based on validation results, choose an appropriate threshold
# We want validation anomaly rate < 1% (ideally < 0.5%)

# Find threshold that gives ~0.5% anomaly rate in validation
desired_val_anomaly_rate = 0.005  # 0.5%

# Sort validation scores and find threshold
sorted_val_scores = np.sort(val_scores_norm)
threshold_idx = int((1 - desired_val_anomaly_rate) * len(sorted_val_scores))
chosen_threshold = sorted_val_scores[threshold_idx]

print(f"=== Threshold Selection ===")
print(f"Desired validation anomaly rate: {desired_val_anomaly_rate*100:.1f}%")
print(f"Chosen threshold: {chosen_threshold:.3f}")

# Apply this threshold to all datasets
train_pred = (train_scores_norm > chosen_threshold).astype(int)
val_pred = (val_scores_norm > chosen_threshold).astype(int)
test_pred_if = (test_scores_norm > chosen_threshold).astype(int)

print(f"\n=== Predictions with threshold {chosen_threshold:.3f} ===")
print(f"Training anomalies: {train_pred.sum()} ({train_pred.mean()*100:.2f}%)")
print(f"Validation anomalies: {val_pred.sum()} ({val_pred.mean()*100:.2f}%)")
print(f"Test anomalies: {test_pred_if.sum()} ({test_pred_if.mean()*100:.2f}%)")
```

```python
# Train additional models for ensemble if needed
# This helps capture different types of anomalies

# Model 2: Elliptic Envelope (good for correlation-breaking anomalies)
print("Training Elliptic Envelope...")
elliptic_env = EllipticEnvelope(
    contamination=0.05,
    random_state=42,
    support_fraction=0.7  # Use 70% of data to estimate covariance
)
elliptic_env.fit(X_train_scaled)

# Get scores (higher = more anomalous)
ee_val_scores = -elliptic_env.decision_function(X_val_scaled)
ee_test_scores = -elliptic_env.decision_function(X_test_scaled)

# Normalize to 0-1
ee_val_scores_norm = score_scaler.fit_transform(ee_val_scores.reshape(-1, 1)).ravel()
ee_test_scores_norm = score_scaler.transform(ee_test_scores.reshape(-1, 1)).ravel()

print("\nElliptic Envelope trained!")
print(f"Val anomaly rate at {chosen_threshold:.3f}: {(ee_val_scores_norm > chosen_threshold).mean()*100:.2f}%")
print(f"Test anomaly rate at {chosen_threshold:.3f}: {(ee_test_scores_norm > chosen_threshold).mean()*100:.2f}%")
```

```python
# Combine predictions from both models
# Weighted average: 60% Isolation Forest, 40% Elliptic Envelope
# Can adjust weights based on validation performance

# Normalize both to same scale first
ensemble_val = 0.6 * val_scores_norm + 0.4 * ee_val_scores_norm
ensemble_test = 0.6 * test_scores_norm + 0.4 * ee_test_scores_norm

# Normalize ensemble scores to 0-1
ensemble_val_norm = score_scaler.fit_transform(ensemble_val.reshape(-1, 1)).ravel()
ensemble_test_norm = score_scaler.transform(ensemble_test.reshape(-1, 1)).ravel()

# Find threshold for ensemble (same 0.5% validation rate)
sorted_ensemble_val = np.sort(ensemble_val_norm)
ensemble_threshold_idx = int((1 - desired_val_anomaly_rate) * len(sorted_ensemble_val))
ensemble_threshold = sorted_ensemble_val[ensemble_threshold_idx]

# Get final ensemble predictions
val_pred_ensemble = (ensemble_val_norm > ensemble_threshold).astype(int)
test_pred_ensemble = (ensemble_test_norm > ensemble_threshold).astype(int)

print(f"=== Ensemble Results ===")
print(f"Ensemble threshold: {ensemble_threshold:.3f}")
print(f"Validation anomalies: {val_pred_ensemble.sum()} ({val_pred_ensemble.mean()*100:.2f}%)")
print(f"Test anomalies: {test_pred_ensemble.sum()} ({test_pred_ensemble.mean()*100:.2f}%)")

print(f"\n=== Comparison: IF vs Ensemble ===")
print(f"Isolation Forest only - Test anomalies: {test_pred_if.sum()} ({test_pred_if.mean()*100:.2f}%)")
print(f"Ensemble (IF+EE) - Test anomalies: {test_pred_ensemble.sum()} ({test_pred_ensemble.mean()*100:.2f}%)")
print(f"Agreement: {(test_pred_if == test_pred_ensemble).mean()*100:.2f}%")
```
