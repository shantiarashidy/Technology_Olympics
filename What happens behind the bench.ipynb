```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, accuracy_score
```

```python
file_path = r'D:\uni and work\Competitions\Data no2\initial\Data\train.csv'
df_train = pd.read_csv(file_path)

df_train.head()
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazirmatn;color:#0099cc">
<font face="vazirmatn" color="#0099cc">
پیش‌پردازش و مهندسی ویژگی
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazirmatn; font-size:medium">
<font face="vazirmatn" size=3>
        در این سوال شما می‌توانید از هر تکنیک پیش‌پردازش/مهندسی ویژگی دلخواهتان، استفاده کنید.
    <br>
    تکنیک‌هایی که استفاده می‌کنید به شکل مستقیم مورد ارزیابی توسط سامانه داوری قرار <b>نمی‌گیرند.</b> بلکه همه آن‌ها در دقت مدل شما تاثیر خواهند گذاشت؛ بنابراین هر چه پیش‌پردازش/مهندسی ویژگی بهتری انجام دهید تا دقت مدل بهبود پیدا کند، امتیاز بیشتری از این سوال کسب خواهید کرد.
    در این قسمت شما می‌توانید بخشی از داده‌ی موجود را برای اعتبارسنجی در نظر بگیرید.
</font>
</p>

```python
categorical_cols = ['StageID', 'GradeID', 'Topic', 'ParentAnsweringSurvey', 
                    'ParentschoolSatisfaction', 'StudentAbsenceDays']
numeric_cols = ['raisedhands', 'VisITedResources', 'AnnouncementsView', 'Discussion']
target_col = 'Class'
```

```python
X_df = df_train[categorical_cols + numeric_cols]
y_df = df_train[target_col]
```

```python
target_encoder = LabelEncoder()
y_encoded = target_encoder.fit_transform(y_df)

# Verify mapping
print("Target mapping:", dict(zip(target_encoder.classes_, target_encoder.transform(target_encoder.classes_))))
```

```python
encoders = {}
X_encoded = X_df.copy()

for col in categorical_cols:
    encoders[col] = LabelEncoder()
    X_encoded[col] = encoders[col].fit_transform(X_df[col])

# Now X_encoded has all columns properly encoded
print("\nEncoded feature shapes:", X_encoded.shape)
print("\nSample of encoded data:")
print(X_encoded.head())
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_encoded, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"\nTraining set: {X_train.shape[0]} samples")
print(f"Validation set: {X_val.shape[0]} samples")
```

<h2 dir=rtl align=right style="line-height:200%;font-family:vazirmatn;color:#0099cc">
<font face="vazirmatn" color="#0099cc">
آموزش مدل
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazirmatn;font-size:medium">
<font face="vazirmatn" size=3>
    حال که داده را پاکسازی کرده و احتمالا ویژگی‌هایی را به آن افزوده یا از آن حذف کرده‌اید، وقت آن است که مدلی آموزش دهید که بتواند متغیر هدف این مسئله را پیش‌بینی کند.
</font>
</p>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazirmatn;font-size:medium">
<font face="vazirmatn" size=3>
توجه داشته باشید که نمونه‌های موجود در مجموعه‌ی آموزش دارای برچسب‌هایی هستند که نشان‌دهنده‌ی سطح هر دانش‌آموز می‌باشند. وظیفه‌ی شما آموزش مدلی است که بتواند با استفاده از ویژگی‌های هر دانش‌آموز، سطح او را برای نمونه‌های موجود در مجموعه‌ی آزمایش پیش‌بینی کرده و به این ترتیب، دانش‌آموزان را دسته‌بندی کند.
</font>
</p>

```python
tree = DecisionTreeClassifier(criterion="entropy", max_depth=4, random_state=42)
tree.fit(X_train, y_train)
```

```python
test_path = r"D:\uni and work\Competitions\Data no2\initial\Data\test.csv"
test = pd.read_csv(test_path)

test.head()
```

```python
y_pred_val = tree.predict(X_val)

print("\n" + "="*50)
print("VALIDATION PERFORMANCE")
print("="*50)
print(f"Accuracy: {accuracy_score(y_val, y_pred_val):.4f}")
print("\nClassification Report:")
print(classification_report(y_val, y_pred_val, target_names=list(target_encoder.classes_)))
```

```python
cv_scores = cross_val_score(tree, X_encoded, y_encoded, cv=5)
print(f"\n5-fold Cross-validation accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
```

```python
feature_importance = pd.DataFrame({
    'feature': categorical_cols + numeric_cols,
    'importance': tree.feature_importances_
}).sort_values('importance', ascending=False)

print("\n" + "="*50)
print("FEATURE IMPORTANCE")
print("="*50)
print(feature_importance)
```

```python
X_test = test[categorical_cols + numeric_cols].copy()
```

```python
for col in categorical_cols:
    # Handle any unseen categories in test data
    # Map unseen categories to -1 (model will still handle it)
    unique_train = set(encoders[col].classes_)
    test_values = test[col].values
    
    # Check for unseen categories
    unseen_mask = ~pd.Series(test_values).isin(unique_train)
    if unseen_mask.any():
        print(f"\n⚠️  Warning: Found unseen categories in '{col}':")
        print(f"    {test_values[unseen_mask]}")
        print("    These will be mapped to -1")
    
    # Transform using existing encoder
    X_test[col] = test[col].map(lambda x: x if x in unique_train else encoders[col].classes_[0])
    X_test[col] = encoders[col].transform(X_test[col])
```

```python
test_predictions = tree.predict(X_test)
```

```python
test_labels = target_encoder.inverse_transform(test_predictions)
```
